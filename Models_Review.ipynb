```python
import seaborn as sns
import matplotlib.pyplot as plt
# sns.set(style="ticks", color_codes=True)
sns.set()
import pandas as pd
```

```python
df = pd.read_csv("model_stats.csv")
```

```python
df = df[['Name', 'Cont', 'Event', 'km2', 'Type', 'HP', 'PP', 'Tot', 'Pes']]
df = df.rename(columns={'Tot':'Processes', 'Pes': 'Pesticides'})
```

```python
df.head()
```

```python
dm = pd.melt(df, 
             id_vars=['Name', 'km2', 'Type', 'HP', 'PP','Processes', 'Pesticides'],
             value_vars=['Cont',
                         'Event'],
             var_name='variable', value_name='Time Scale')
# 
```
