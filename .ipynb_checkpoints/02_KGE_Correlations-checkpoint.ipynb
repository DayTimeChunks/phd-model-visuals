# A

```python
g = sns.lmplot(data=ff, x='KGE-CONC_out', y='KGE-CONC-tra', hue='Model')
```

Good outlet concentrations at this point are not dominating the distribution of models tested. 

# B

```python
g = sns.lmplot(data=ff, x='KGE-CONC_out', y='KGE-Q_out', hue='Model')
```

Supports idea that good concentrations at the outlet go hand in hand with good discharge. Therefore, we should use Q-metrics as constraints, for either model. 

This also shows, that the fixed model is not doing as well in predicing export at this point. 

# C

```python
g = sns.lmplot(data=ff, x='KGE-CONC-tra', y='KGE-Q_out', hue='Model')
```

As small correlation for var model (and none for fix), potentially indicates that soil moisture constraints may be correct. 

## Take home message

- Outlet is a necessary constraint because:
    - Based on outlet discharge we can be more certain that moisture in soils is well represented, which is why soil concentrations show positive correlations to discharge AND outlet concentrations on "var" model. 
    - Outlet concentrations, should also provide an indication of the available pools, and thus provide support of the model's value, even if not specialized for this purpose.  

- Limitations
The model is already showing its real limitations with respect to representing catchment offsite transport, as it is not taking into account the key processes known to be of key importance (e.g. preferential flow as well as rainfall run-off processes, for which event-based models are specialized in).
